```python
image = img_to_array(image)
type(image),image.shape
```

```python
input_image = image.reshape((1,)+image.shape)
input_image.shape
```

```python
data_gen = ImageDataGenerator(rotation_range=30, width_shift_range=0.1,
                              height_shift_range=0.3, shear_range=0.1,
                              zoom_range=0.2,fill_mode='nearest',
                              horizontal_flip=True, vertical_flip=False,  data_format=None,
                              validation_split=0.0, interpolation_order=1, dtype=None)
```

```python
i = 0
for output in data_gen.flow(input_image,batch_size = 1,save_to_dir =r'/content/drive/MyDrive/Portfolio/100 DAYS OF DL/Data Augmentation/aug_data'):
  i += 1
  if i == 10:
    break
```
