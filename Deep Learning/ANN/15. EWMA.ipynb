```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/content/drive/MyDrive/Portfolio/100 DAYS OF DL/DailyDelhiClimateTrain.csv')
df.head()
```

```python
df = df[['date','meantemp']]
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['date'], df['meantemp'])
```

```python
x1 = df['meantemp'].ewm(alpha=0.1).mean()
```

```python
df['ewma']=x1
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['date'], df['meantemp'])
plt.plot(df['date'],df['ewma'],color = 'red')
```

```python
for i in (np.arange(0.1,1,0.1)):
  x1 = df['meantemp'].ewm(alpha = float(i)).mean()
  plt.figure(figsize=(10,5))
  plt.scatter(df['date'], df['meantemp'])
  plt.plot(df['date'],x1,color = 'red')
  plt.title(f'Alpha = {i}')
  plt.show()
```
