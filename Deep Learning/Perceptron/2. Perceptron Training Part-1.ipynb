```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
sns.set()
```

```python
from sklearn.datasets import make_classification
```

```python
Sample_Size = 100
Max_Epoch = 200
```

```python
X,y = make_classification(n_samples=Sample_Size,
    n_features=2,
    n_informative=1,
    n_redundant=0,
    n_repeated=0,
    n_classes=2,
    n_clusters_per_class=1,
    class_sep=1.0,
    hypercube=False,
    random_state=42)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c = y, cmap = 'winter')
plt.show()
```

```python
def perceptron(X,y):
    X = np.insert(X,0,1,axis = 1)    #Adding a 
    weights = np.ones(X.shape[1])
    lr = 0.01
    for i in range(Max_Epoch):
         j = np.random.randint(0,Sample_Size)
         y_hat = step(np.dot(X[j],weights))
         weights = weights + lr * (y[j] - y_hat) * X[j]
    return weights[0],weights[1:]
```

```python
def step(z):
    if z > 0:
        return 1
    else:
        return 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(intercept_)
print(coef_)
```

```python
m = -coef_[0]/coef_[1]
c = -intercept_/coef_[1]
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input+c
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap = 'winter')
plt.plot(x_input,y_input)
plt.show()
```
