```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
sns.set()
```

```python
from sklearn.datasets import make_classification
X,y = X,y = make_classification(n_samples=200,
    n_features=2,
    n_informative=1,
    n_redundant=0,
    n_repeated=0,
    n_classes=2,
    n_clusters_per_class=1,
    class_sep=1.0,
    hypercube=False,
    random_state=42)
```

```python
X.shape
```

```python
y.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap ='winter')
plt.show()
```

```python
def perceptron_training_using_loss_fn(X,y):
    w1=w2=b=1
    lr = 0.1
    epoch = 1000
    for i in range(epoch):
        for j in range(X.shape[0]):
            z = w1*X[j][0]+w2*X[j][1]+b
            if z*y[j] < 0:
                w1 = w1 + lr*y[j]*X[j][0]
                w2 = w2 + lr*y[j]*X[j][1]
                b = b + lr*y[j]
    return w1,w2,b
```

```python
w1,w2,b = perceptron_training_using_loss_fn(X,y)
```

```python
b,w1,w2
```

```python
x_line = np.linspace(-2,2,10000)
y_line = (b/w2)+(w1/w2)*x_line
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap ='winter')
plt.plot(x_line,y_line)
plt.show()
```
