```python
import pandas as pd
import requests
```

```python
url = "https://api.coindesk.com/v1/bpi/currentprice.json"

response = requests.get(url)

print(response.text)
```

```python
data = response.json()
```

```python
df = pd.DataFrame(data)
df
```
