##### Flipping Image

###### Used for Data Augmentation to avoid over fitting in CNN

```python
img = cv2.imread(path2)
cv2.imshow('img',img)
cv2.waitKey(0)
print(img.shape)
plt.imshow(img)
```

###### Flipping UD

```python
img_ud = cv2.flip(img,0)
cv2.imshow('img_ud',img_ud)
cv2.waitKey(0)
print(img_ud.shape)
plt.imshow(img_ud)
```

###### Flipping Side

```python
img_sd = cv2.flip(img,1)
cv2.imshow('img_sd',img_sd)
cv2.waitKey(0)
print(img_sd.shape)
plt.imshow(img_sd)
```

```python
img_sd_ud = cv2.flip(img,-1)
cv2.imshow('img_sd_ud',img_sd_ud)
cv2.waitKey(0)
print(img_sd_ud.shape)
plt.imshow(img_sd_ud)
```

```python
img_flip = cv2.resize(np.hstack((img,img_ud,img_sd,img_sd_ud)), (1000,500))
cv2.imshow('comparison',img_flip )
cv2.waitKey(0)
plt.imshow(img_flip)
plt.show()
```

##### Creating Shapes

```python
own_image = np.zeros((512,512,3))
cv2.imshow('own_img',own_image)
cv2.waitKey(0)
```

###### Rectangle

```python
cv2.rectangle(own_image,pt1=(100,100),pt2=(400,400),color= (0,0,255),thickness = 5) ### RedLine (B,G,R)
cv2.imshow('own_img',own_image)
cv2.waitKey(0)
```

###### Rectangular Area (Thickness = -1)

```python
cv2.rectangle(own_image,pt1=(100,100),pt2=(400,400),color= (0,100,0),thickness = -1) ### Green BoxFill (B,G,R)
cv2.imshow('own_img',own_image)
cv2.waitKey(0)
```

###### Circle

```python
cv2.circle(own_image, center=(250,250),radius = 100,color =  (10,0,0),thickness = -1) ### Blue Circle (B,G,R)
cv2.imshow('own_img',own_image)
cv2.waitKey(0)
```

###### Line

```python
cv2.line(own_image,pt1=(0,0),pt2 =(512,512),color = (10,100,200),thickness = 4) ### Blue Circle (B,G,R)
cv2.imshow('own_img',own_image)
cv2.waitKey(0)
```

##### Adding Text

```python
image_up = cv2.putText(own_image,org = (100,50), fontScale =2, color = (0,255,255),lineType = cv2.LINE_AA,text = "It's SPD", fontFace= cv2.FONT_ITALIC)
cv2.imshow('tected image',image_up)
cv2.waitKey(0)
```
