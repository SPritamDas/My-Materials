The reason for using **loan default prediction** literature for a **loan approval prediction** project is that both tasks involve binary classification. The choice of train-test split ratio can significantly impact the performance of your machine learning models. A common guideline is to use a split of 70-30, 80-20, or even 90-10. Here‚Äôs a breakdown of these ratios:

1. 70-30 Split:

Pros: This is a balanced split that provides a sufficient amount of data for both training and testing. It allows the model to learn adequately while still having enough data to evaluate its performance.
Cons: The training set may be slightly smaller, which could be a concern for very complex models or when dealing with limited data.<br>

2. 80-20 Split:

Pros: This is one of the most commonly used ratios. It provides a good balance, allowing for robust training while maintaining a significant testing set for evaluation.
Cons: The test set might not be as large, which could affect the reliability of the performance metrics if the data has high variance.<br>

3. 90-10 Split:

Pros: More data for training can lead to better model performance, especially for complex models or when data is scarce.
Cons: A smaller test set might lead to unreliable performance metrics, particularly if the test data has a different distribution than the training data.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from keras_tuner import RandomSearch
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier, HistGradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.svm import SVC

# Initialize base models
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
lgb_model = LGBMClassifier()
catboost_model = CatBoostClassifier(silent=True)
gb_model = HistGradientBoostingClassifier()

# Define a function for creating the ANN model for Keras Tuner
def build_ann_model(hp):
    model = Sequential()
    model.add(Dense(units=hp.Int('units_input', min_value=32, max_value=256, step=32), activation='relu', input_shape=(x_train.shape[1],)))
    model.add(Dropout(rate=hp.Float('dropout_input', min_value=0.0, max_value=0.5, step=0.1)))
    
    # Additional hidden layers
    for i in range(hp.Int('num_layers', 1, 3)):
        model.add(Dense(units=hp.Int(f'units_{i}', min_value=32, max_value=256, step=32), activation='relu'))
        model.add(Dropout(rate=hp.Float(f'dropout_{i}', min_value=0.0, max_value=0.5, step=0.1)))
    
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
    return model

# Set up Keras Tuner for hyperparameter optimization
tuner = RandomSearch(
    build_ann_model,
    objective='val_auc',
    max_trials=10,
    executions_per_trial=1,
    directory='ann_tuning',
    project_name='stacking_ann'
)

# Wrap the ANN model with Keras Tuner
ann_meta_model = KerasClassifier(build_fn=tuner.hypermodel.build, epochs=20, batch_size=32, verbose=0)

# Initialize Stacking Classifier with optimized ANN model as meta-model
stacking_clf = StackingClassifier(
    estimators=[
        ('xgb', xgb_model),
        ('lgb', lgb_model),
        ('cat', catboost_model),
        ('gb', gb_model),
    ],
    final_estimator=ann_meta_model,
    stack_method='predict_proba',
    n_jobs=-1
)

# Search for the best hyperparameters
tuner.search(x_train, y_train, validation_data=(x_val, y_val), epochs=20, batch_size=32)

# Retrieve the best ANN model after tuning
best_hyperparameters = tuner.get_best_hyperparameters(1)[0]
best_ann_model = tuner.hypermodel.build(best_hyperparameters)

# Fit the stacking classifier with the optimized ANN as the meta-model
stacking_clf.final_estimator_ = best_ann_model
stacking_clf.fit(x_train, y_train)

# Make probability predictions on training and validation sets
y_train_pred_stack = stacking_clf.predict_proba(x_train)[:, 1]
y_val_pred_stack = stacking_clf.predict_proba(x_val)[:, 1]

# Calculate AUC for training and validation sets
train_auc = roc_auc_score(y_train, y_train_pred_stack)
val_auc = roc_auc_score(y_val, y_val_pred_stack)

# Print out the AUC scores for evaluation
print(f"Training AUC: {train_auc:.4f}")
print(f"Validation AUC: {val_auc:.4f}")

# Optionally, print out the first few predictions for verification
print("Training Predictions (Probabilities):", y_train_pred_stack[:5])
print("Validation Predictions (Probabilities):", y_val_pred_stack[:5])
```
